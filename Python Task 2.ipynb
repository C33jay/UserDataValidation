```python
#import libraries
import string
import random

#obtain user details
def obtain_details():
    first_name = input("Enter your first name: ")
    second_name = input("Enter your second name: ")
    email = input("Enter your email address: ")
    details =[first_name, second_name, email,]
    
    return details

#generates password 
def gen_password(details):
    characters = string.ascii_lowercase
    length = 5
    random_password = ''.join(random.choice(characters)for i in range (length))
    
    password = str(details[0][0:2] +  details[1][-2:] + random_password)
    
    return password

#program proper
status = True
users = {}
user_number = 1

while status:
    
    #obtains user details
    details = obtain_details()
    users.update({user_number : details,})
    
    #Displays generated password
    password = gen_password(details)
    print("Your password is: " + password)
   
    
    satisfied_with_password = input("Are you satisfied with your password? [Enter Yes or No]")
    
    password_loop = True
    

    while password_loop:
        if satisfied_with_password == "Yes":
            #Adds genetrated password to user details
            details.append(password)
                      
            password_loop = False
            
        else:
            password =input("Enter your password: ")
            
            pass_len = True
            while pass_len:
                if len(password)>=7: 
                    #Adds choosen password to user details if it meets the condition
                    details.append( password)
                    
                    
                    
                else:
                    print("Your password should not be less than 7")
                    #Gives user another chance to input a password that meets the condition
                    password = input("Enter another password: ")
                    
                    if len(password) >= 7:
                        #Adds choosen password to user details if it meets the condition
                        details.append(password)
                        password_loop = False
                        pass_len = False
                    
                    
                    
                                  
    #requests for new user       
    new_user = input("Any other user? [Yes or No]")
    
    if (new_user == "No"):
        status = False
        
        #prints users with their details
        for k,v in users.items():
            for value in[k]:
                print(k)
                for value_ in v:
                    print(value_)
         
                
    else:
        
        status = True
        user_number = user_number + 1
        
        
        
        
```
